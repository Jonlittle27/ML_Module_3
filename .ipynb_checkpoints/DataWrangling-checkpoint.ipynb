#### Step 4 – Examine the data for missing observations and duplicate records. Suggest what methods are appropriate to handle them.

```python
#Perform the action(s) in step 4 to handle any missing data and duplicate records that were found.

#finding missing data counts
data.isna().sum()
```

```python
#there are multiple columns with missing values but the amount is relatively low to the size of the data set
#so dropping the rows seems to be the best course of action to ensure accurate data
newdata = data.copy()
newdata = newdata.drop("ehail_fee", axis = 1)
newdata = newdata.dropna()
newdata.isna().sum()
```

```python
#Print the number of rows and columns in data.
print("Number of Rows")
print(newdata.shape[0])
print("---"*5)
print("Number of Columns")
print(newdata.shape[1])
```

```python
#checking for duplicate records
newdata.duplicated().sum()
```

```python
#dropping duplicate record
newdata = newdata.drop_duplicates()

newdata.duplicated().sum()
```

#### Step 5 – Plot a histogram showing the distribution of the trip distance. Discuss your observations (no more than 1 paragraph).

```python
#histogram plot
n, bins, patches = plt.hist(x=newdata["trip_distance"], bins='auto',
                            alpha=0.7)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Distance')
plt.ylabel('Frequency')
plt.title('Trip Distance Distribution')

# My main observation is that there are clearly outliers in this data that are vastly skewing the size and scope of 
# the histogram.  With the bins set to auto it should be generating an easy to view chart.  But with so much of the data 
# being focused in the very small trip distances, probably due to taxi rides being relatively short in just New York City, it
# allows the outliers to throw the chart off completely.  
```

#### Step 6 – Did you detect any outliers in step 5? If so, perform a suitable method to handle them?

```python
#It is clear by the scope of the chart that there are outliers and I think filtering them out is the best method

clean_data = newdata.copy()
clean_data = clean_data[clean_data["trip_distance"] >= 0]
clean_data = clean_data[clean_data["trip_distance"] <= 30]

#re-visualizing the histogram
n, bins, patches = plt.hist(x=clean_data["trip_distance"], bins='auto',
                            alpha=0.7)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Distance')
plt.ylabel('Frequency')
plt.title('Trip Distance Distribution')
```
